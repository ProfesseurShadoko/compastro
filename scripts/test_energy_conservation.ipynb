```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import os
while not os.path.exists('files'):
    os.chdir('..')
print("Current Working Directory:", os.getcwd())
```

```python
def get_df(file):
    df = pd.read_csv(file)
    df_time = df.groupby('t').mean().reset_index()[["t","envEnergy"]]
    return df_time
```

```python
df_direct = get_df("files/output/direct_energy_conservation.csv")
df_mono = get_df("files/output/mono_energy_conservation.csv")
df_quad = get_df("files/output/quad_energy_conservation.csv")
```

```python
plt.plot(df_direct["t"], df_direct["envEnergy"], label="Direct")
plt.plot(df_mono["t"], df_mono["envEnergy"], label="Mono")
plt.plot(df_quad["t"], df_quad["envEnergy"], label="Quad")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```
